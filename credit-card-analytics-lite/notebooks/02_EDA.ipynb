```python
import pandas as pd
import numpy as np

import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
import os

warnings.filterwarnings('ignore')

# ==========================================
# Create plots folder
# ==========================================

os.makedirs('../plots', exist_ok=True)

# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv('../data/credit_card_data.csv')

print("Dataset Loaded Successfully")
print(df.shape)

# ==========================================
# Create Figure
# ==========================================

fig, axes = plt.subplots(
    2,
    2,
    figsize=(14, 10)
)

fig.suptitle(
    'Credit Card Analytics — EDA',
    fontsize=16,
    fontweight='bold'
)

# ==========================================
# Chart 1 — Monthly Spend Distribution
# ==========================================

axes[0,0].hist(
    df['monthly_spend'],
    bins=40,
    color='#2563EB',
    edgecolor='white',
    alpha=0.85
)

axes[0,0].axvline(
    df['monthly_spend'].median(),
    color='red',
    linestyle='--',
    linewidth=2
)

axes[0,0].set_title(
    'Monthly Spend Distribution',
    fontweight='bold'
)

axes[0,0].set_xlabel(
    'Monthly Spend'
)

# ==========================================
# Chart 2 — Payment Behavior
# ==========================================

pb = df['payment_behavior'].value_counts()

axes[0,1].bar(
    pb.index,
    pb.values,
    color=[
        '#16A34A',
        '#3B82F6',
        '#F59E0B',
        '#EF4444'
    ],
    edgecolor='black'
)

axes[0,1].set_title(
    'Payment Behavior',
    fontweight='bold'
)

axes[0,1].set_ylabel('Customers')

# ==========================================
# Chart 3 — Utilization Distribution
# ==========================================

axes[1,0].hist(
    df['util_rate'],
    bins=40,
    color='#7C3AED',
    edgecolor='white',
    alpha=0.85
)

axes[1,0].axvline(
    0.30,
    color='green',
    linestyle='--',
    linewidth=2,
    label='Healthy'
)

axes[1,0].axvline(
    0.80,
    color='red',
    linestyle='--',
    linewidth=2,
    label='Risk'
)

axes[1,0].legend()

axes[1,0].set_title(
    'Credit Utilization',
    fontweight='bold'
)

# ==========================================
# Chart 4 — Spend vs Credit Limit
# ==========================================

sample = df.sample(2000)

axes[1,1].scatter(
    sample['credit_limit']/1000,
    sample['monthly_spend']/1000,
    alpha=0.3,
    color='#2563EB',
    s=10
)

axes[1,1].set_title(
    'Spend vs Credit Limit',
    fontweight='bold'
)

axes[1,1].set_xlabel(
    'Credit Limit (Thousands)'
)

axes[1,1].set_ylabel(
    'Monthly Spend (Thousands)'
)

# ==========================================
# Save Plot
# ==========================================

plt.tight_layout()

plt.savefig(
    '../plots/eda_dashboard.png',
    dpi=100,
    bbox_inches='tight'
)

plt.close()

# ==========================================
# Business Insights
# ==========================================

print()
print("KEY BUSINESS INSIGHTS")
print("-" * 40)

high_util = df[
    df['util_rate'] > 0.80
]['monthly_spend'].mean()

low_util = df[
    df['util_rate'] < 0.30
]['monthly_spend'].mean()

print(
    f"1. High utilization users spend Rs {high_util:,.0f}/month"
)

print(
    f"2. Low utilization users spend Rs {low_util:,.0f}/month"
)

full_pay = df[
    df['payment_behavior'] == 'Full'
]['util_rate'].mean() * 100

none_pay = df[
    df['payment_behavior'] == 'None'
]['util_rate'].mean() * 100

print(
    f"3. Full payers utilization: {full_pay:.1f}%"
)

print(
    f"4. Non-payers utilization: {none_pay:.1f}%"
)

tier_spend = df.groupby(
    'city_tier'
)['monthly_spend'].mean()

print()
print("Average Spend by City Tier:")

for tier, spend in tier_spend.items():

    print(
        f"{tier}: Rs {spend:,.0f}"
    )

print()
print("Saved:")
print("plots/eda_dashboard.png")
```
