```python
import pandas as pd
import numpy as np
import os

# ==========================================
# Create data folder
# ==========================================

os.makedirs('../data', exist_ok=True)

# ==========================================
# Random seed
# ==========================================

np.random.seed(42)

# ==========================================
# Dataset Size (LIGHT VERSION)
# ==========================================

N = 30000

print("Generating lightweight credit card dataset...")

# ==========================================
# Customer Profile
# ==========================================

age = np.random.randint(22, 65, N)

income_mo = np.random.lognormal(
    mean=10.5,
    sigma=0.6,
    size=N
).round(0)

city_tier = np.random.choice(
    ['Tier1', 'Tier2', 'Tier3'],
    N,
    p=[0.35, 0.40, 0.25]
)

cibil = np.clip(
    np.random.normal(710, 70, N),
    300,
    900
).astype(int)

tenure_mo = np.random.randint(
    1,
    84,
    N
)

# ==========================================
# Credit Limit
# ==========================================

base_limit = (
    income_mo *
    np.random.uniform(1.5, 3.5, N)
).clip(10000, 500000)

credit_limit = (
    np.round(base_limit, -3)
).astype(int)

# ==========================================
# Monthly Spending
# ==========================================

monthly_spend = (
    credit_limit *
    np.random.uniform(0.15, 0.90, N)
).round(0)

food_pct = np.random.uniform(0.10, 0.35, N)

travel_pct = np.random.uniform(0.05, 0.25, N)

shopping_pct = np.random.uniform(0.10, 0.40, N)

entertainment_pct = np.random.uniform(0.03, 0.15, N)

utilities_pct = (
    1 -
    food_pct -
    travel_pct -
    shopping_pct -
    entertainment_pct
).clip(0.05, 0.30)

food_spend = (
    monthly_spend * food_pct
).round(0)

travel_spend = (
    monthly_spend * travel_pct
).round(0)

shopping_spend = (
    monthly_spend * shopping_pct
).round(0)

entertainment_spend = (
    monthly_spend * entertainment_pct
).round(0)

utilities_spend = (
    monthly_spend * utilities_pct
).round(0)

# ==========================================
# Behavioral Features
# ==========================================

txn_count = np.random.poisson(
    16,
    N
).clip(1, 60)

avg_txn_value = (
    monthly_spend / txn_count
).round(0)

util_rate = (
    monthly_spend / credit_limit
).round(3).clip(0.01, 0.99)

payment_behavior = np.random.choice(
    ['Full', 'Partial', 'Minimum', 'None'],
    N,
    p=[0.45, 0.30, 0.17, 0.08]
)

emi_conversion = np.random.binomial(
    1,
    0.25,
    N
)

reward_pts_mo = np.random.poisson(
    700,
    N
).clip(0, 3000)

intl_txn_pct = np.random.exponential(
    0.06,
    N
).clip(0, 0.50).round(3)

missed_payments = np.random.poisson(
    0.2,
    N
).clip(0, 5)

autopay_enabled = np.random.binomial(
    1,
    0.65,
    N
)

# ==========================================
# Create DataFrame
# ==========================================

df = pd.DataFrame({

    'cust_id': [
        f"CC{i:07d}" for i in range(N)
    ],

    'age': age,

    'income_mo': income_mo.astype(int),

    'city_tier': city_tier,

    'cibil': cibil,

    'tenure_mo': tenure_mo,

    'credit_limit': credit_limit,

    'monthly_spend': monthly_spend.astype(int),

    'food_spend': food_spend.astype(int),

    'travel_spend': travel_spend.astype(int),

    'shopping_spend': shopping_spend.astype(int),

    'entertainment_spend':
        entertainment_spend.astype(int),

    'utilities_spend':
        utilities_spend.astype(int),

    'txn_count': txn_count,

    'avg_txn_value':
        avg_txn_value.astype(int),

    'util_rate': util_rate,

    'payment_behavior':
        payment_behavior,

    'emi_conversion':
        emi_conversion,

    'reward_pts_mo':
        reward_pts_mo,

    'intl_txn_pct':
        intl_txn_pct,

    'missed_payments':
        missed_payments,

    'autopay_enabled':
        autopay_enabled,
})

# ==========================================
# Optimize Memory
# ==========================================

for col in [
    'age',
    'cibil',
    'txn_count',
    'reward_pts_mo',
    'missed_payments'
]:
    df[col] = df[col].astype('int16')

# ==========================================
# Save CSV
# ==========================================

df.to_csv(
    '../data/credit_card_data.csv',
    index=False
)

# ==========================================
# Output
# ==========================================

print()
print(f"Shape: {df.shape}")

print(
    f"Average Monthly Spend: Rs {monthly_spend.mean():,.0f}"
)

print(
    f"Average Credit Limit: Rs {credit_limit.mean():,.0f}"
)

print(
    f"Average Utilization: {util_rate.mean()*100:.1f}%"
)

print()
print("Payment Behavior Distribution:")

for b, c in zip(
    *np.unique(payment_behavior, return_counts=True)
):

    print(
        f"{b:10}: {c:,} ({c/N*100:.1f}%)"
    )

print()
print("Saved:")
print("data/credit_card_data.csv")
```
